**Please keep in mind that these temperature values were obtained from the internet and may differ somewhat from the actual temperature for that year. So the model being taught below may not be perfect.**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
tempdf = pd.read_csv('averageTemperatures.csv')
```

```python
tempdf.head()
```

```python
tempdf['averageCelsiusTemperature'] = (tempdf['Average_Fahrenheit_Temperature'] - 32) * 5/9
tempdf.drop('Average_Fahrenheit_Temperature', axis=1, inplace=True)
tempdf.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(tempdf['Year'], tempdf['averageCelsiusTemperature'])
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.title('Temperature Change from 1900 to 2023')
plt.show()
```

```python
X = tempdf['Year'].values.reshape(-1, 1)
y = tempdf['averageCelsiusTemperature'].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)


print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R-squared:', r2_score(y_test, y_pred))

plt.figure(figsize=(12, 6))
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.title('Temperature Change from 1900 to 2023')
plt.show()
```

```python
year2020 = np.array([[2020]])
predictedTemperatureOf2020 = regressor.predict(year2020)
print(f"The predicted average temperature for the year 2020 is: {predictedTemperatureOf2020[0][0]}°C")
```

```python
year2020FromCSV = tempdf[tempdf['Year'] == 2020]['averageCelsiusTemperature'].values[0]
print(f"The actual average temperature for the year 2020 was: {year2020FromCSV}°C")
```
